```python
with open('douglas_data.csv') as f:
    r = csv.reader(f)
    data = list(r)


bend_strength = [ float(r[-1]) for r in data[2:]]
knot_ratio = [ float(r[2]) for r in data[2:]]

bend_strength_Nm2 = [1e6 * e for e in bend_strength]


plt.plot(knot_ratio, bend_strength_Nm2, 'o')
plt.xlabel('Knot ratio')
plt.ylabel('Bend strength Nm2')
plt.show()
```

# Exercise 4

**Question 1**

```python
A = 1e-4
E = 1e6
alpha = 0.08

N = 100
lam = [1 + 9 * n / 99 for n in range(0, 100)]

F_H = [3 * E * A * (e - 1) for e in lam]
F_nH = [E * A  * (e - 1 / e**2) for e in lam]
F_G = [E * A * (e - 1 / e**2) / (1 - alpha * (e - 1)) for e in lam]

plt.plot(lam, F_H, lam, F_nH, lam, F_G)
plt.legend(["Hookean", "neo-Hookean", "Gent"])
plt.xlabel("$\lambda$ (-)")
plt.ylabel("Force (N)")
plt.show()
```

**Question 2**

```python
with open('force_stretch.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerow(["Stretch (-)", "Hooke (N)", "neo-Hookean (N)", "Gent (N)"])
    for n in range(N):
        writer.writerow([lam[n], F_H[n], F_nH[n], F_G[n]])
```
