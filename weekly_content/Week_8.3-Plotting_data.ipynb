# Getting started

To import `matplotlib.pyplot` it is common to use the code

```python
import matplotlib.pyplot as plt
```

This should be placed at the top of your Python file

The `as` keyword is used to create our own, shorter name for `matplotlib.pyplot`; in this case, `plt`

# Loading some data to plot

Let's load in the temperature data so we can plot it

```python
import csv
with open('sample_data/temperature.csv', 'r') as file:
    reader = csv.reader(file)
    data = list(reader)
    
print(data)
```

We now need to extract the time and temperature values and turn them into floats

```python
N_rows = len(data)

time = [float(data[n][0]) for n in range(1, N_rows)]
temp = [float(data[n][1]) for n in range(1, N_rows)]
```

# Creating a line plot

To plot this data as a curve, we use the `plot` and `show` functions as follows

```python
plt.plot(time, temp)
plt.show()
```

# Spicing up the figure

* The plot that is produced is very basic and some important information is missing
    * A label for the $x$ axis
    * A label for the $y$ axis
* What if we want to change the colour, style, and thickness of the line?

```python
plt.plot(time, temp, 'r', linewidth = 2)
plt.xlabel('Time of day')
plt.ylabel('Temperature (C)')
plt.show()
```

# Plotting as data points

Plotting the data points rather than the line is simple:

```python
plt.plot(time, temp, 'r.', markersize = 8)
plt.xlabel('Time of day')
plt.ylabel('Temperature (C)')
plt.show()
```

# Saving your plot

The `savefig` function can be used to save your plot in different formats

```python
# Save the figure as a PDF
plt.plot(time, temp, 'r.', markersize = 8)
plt.xlabel('Time of day')
plt.ylabel('Temperature (C)')

# Save the figure as a PDF then PNG
plt.savefig('temperature_vs_time.pdf')
plt.savefig('temperature_vs_time.png')
```
